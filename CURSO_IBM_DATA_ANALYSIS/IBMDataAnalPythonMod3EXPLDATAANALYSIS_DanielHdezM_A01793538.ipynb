## CORRELACIONAMIENTO Y CAUSALIDAD##

Aqui es importante diferenciar entre ambos conceptos, correlacionamiento significa que hay una relación  entre una y otra variable, pero no habria una dependencia, y causalidad significaria que hay una relación de causa y efecto entre ellas, por lo que es importante tener claro que correlación no implica causalidad.

**Correlacion de Pearson**

La función de correlacion de Pearson mide la dependencia líneal dentreo dos variables, el resultado de esta correlacion va de -1 a 1 en donde si el valor es cercano a cero, se infiere que hay poca correlación entre las variables:

1: Perfect positive linear correlation.
0: No linear correlation, the two variables most likely do not affect each other.
-1: Perfect negative linear correlation.

**P-VALUE**

Es el valor de probabilidad de que la correlación entre dos variables es estadisiticamente significativa. Generalmente unn valor mejor a 0.05 significa que hay una fuerte correlacion entre dichas variables.

p-value is  
<
  0.001: we say there is strong evidence that the correlation is significant.
the p-value is  
<
  0.05: there is moderate evidence that the correlation is significant.
the p-value is  
<
  0.1: there is weak evidence that the correlation is significant.
the p-value is  
>
  0.1: there is no evidence that the correlation is significant.

##**ANOVA**##
##ANALISIS DE VARIANZA##

Este es un método estadistico usado para probar si hay una varianza entre las medias de dos o mas grupos. ANOVA arroja dos parámetros:

* F-test score: ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score. A larger score means there is a larger difference between the means.

* P-value: P-value tells how statistically significant our calculated score value is.

En nuestro caso de estudio, si price esta fuerte correlaciondo con las variables que analizaremos ( drive-wheels & body-style") los valores que arrojará la función de ANOVA serán considerable para el F-test y P-value cercano a cero.


```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```
