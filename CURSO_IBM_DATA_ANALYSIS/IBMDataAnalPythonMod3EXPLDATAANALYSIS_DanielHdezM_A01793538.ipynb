## GROUPBY##

A continuaci√≥n haremos un ejemplo con las columnas de drive.wheels, body.style y price.

```python
df['drive-wheels'].unique()
```

Es importante que usemos un dataset pivote, para que podamos integrar dichas variables sin afectar los datos originales

```python
df_group_one = df[['drive-wheels','body-style','price']]
```

```python
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

Como podemos ver en el siguiente heatmap, no permite ver como las caracteristicas de tracci√≥n y body-style influyen directamente en el precio. rapidamente podemos tener una referencia de las variables que pueden ser buenos predictores

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

## CORRELACIONAMIENTO Y CAUSALIDAD##

Aqui es importante diferenciar entre ambos conceptos, correlacionamiento significa que hay una relaci√≥n  entre una y otra variable, pero no habria una dependencia, y causalidad significaria que hay una relaci√≥n de causa y efecto entre ellas, por lo que es importante tener claro que correlaci√≥n no implica causalidad.

**Correlacion de Pearson**

La funci√≥n de correlacion de Pearson mide la dependencia l√≠neal dentreo dos variables, el resultado de esta correlacion va de -1 a 1 en donde si el valor es cercano a cero, se infiere que hay poca correlaci√≥n entre las variables:

1: Perfect positive linear correlation.
0: No linear correlation, the two variables most likely do not affect each other.
-1: Perfect negative linear correlation.

**P-VALUE**

Es el valor de probabilidad de que la correlaci√≥n entre dos variables es estadisiticamente significativa. Generalmente unn valor mejor a 0.05 significa que hay una fuerte correlacion entre dichas variables.

p-value is  
<
  0.001: we say there is strong evidence that the correlation is significant.
the p-value is  
<
  0.05: there is moderate evidence that the correlation is significant.
the p-value is  
<
  0.1: there is weak evidence that the correlation is significant.
the p-value is  
>
  0.1: there is no evidence that the correlation is significant.

##**ANOVA**##
##ANALISIS DE VARIANZA##

Este es un m√©todo estadistico usado para probar si hay una varianza entre las medias de dos o mas grupos. ANOVA arroja dos par√°metros:

* F-test score: ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score. A larger score means there is a larger difference between the means.

* P-value: P-value tells how statistically significant our calculated score value is.

En nuestro caso de estudio, si price esta fuerte correlaciondo con las variables que analizaremos ( drive-wheels & body-style") los valores que arrojar√° la funci√≥n de ANOVA ser√°n considerable para el F-test y P-value cercano a cero.


```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```
